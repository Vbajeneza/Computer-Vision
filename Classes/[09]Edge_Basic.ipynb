```python
import numpy as np
import cv2
import matplotlib.pylab as plt

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")    
```

```python
src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

mx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype=np.float32)
my = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]], dtype=np.float32)

dx = cv2.filter2D(src, -1, mx, delta=128)
dy = cv2.filter2D(src, -1, my, delta=128)

imshow('src', src); plt.show()
imshow('dx', dx); plt.show()
imshow('dy', dy); plt.show()
```

```python
dx = cv2.Sobel(src, cv2.CV_32F, 1, 0)
dy = cv2.Sobel(src, cv2.CV_32F, 0, 1)

fmag = cv2.magnitude(dx, dy)
mag = np.uint8(np.clip(fmag, 0, 255))#to change pixel ito range 0~255 by clipping function,less than 0 become 0. gt than 255 become255
_, edge = cv2.threshold(mag, 150, 255, cv2.THRESH_BINARY)

imshow('mag', mag); plt.show()
imshow('edge', edge); plt.show()
```

```python
gradient_angle = cv2.phase(dx, dy, angleInDegrees=True)
print(np.min(gradient_angle), np.max(gradient_angle))
gradient_angle
```

```python
imshow('angle', gradient_angle); plt.show()
```
